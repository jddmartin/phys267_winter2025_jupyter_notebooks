```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
```

```python
masses_g = np.array([200, 300, 400, 500, 600, 700, 800, 900])
extensions_cm = np.array([5.1, 5.5, 5.9, 6.8, 7.4, 7.5, 8.6, 9.4,])
N = len(masses_g)
sigma = 0.2
lr = ss.linregress(masses_g, extensions_cm)
```

```python
ls = np.linspace(200, 900, 10)
plt.errorbar(masses_g, extensions_cm, yerr=0.2, xerr=None, fmt="none")
ax = plt.gca()
ax.plot(masses_g, extensions_cm, "+")
ax.plot(ls, lr.slope*ls + lr.intercept)
ax.set_ylabel("extension (cm)")
ax.set_xlabel("mass (g)")
```

```python
chi2_obs = sum((masses_g*lr.slope+lr.intercept - extensions_cm)**2)/sigma**2
```

```python
df = N-2
```

```python
a = ss.chi2.cdf(chi2_obs, df=df)
```

```python
p = 1-a
print(p)  # probability of observing a chi2 of observed value or higher under assumptions
```

```python
# observed p value is not small enough to reject hypothesis.  We expect to see a p-value this small one in five times.
```
