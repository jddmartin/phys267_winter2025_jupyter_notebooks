```python
# Test of variance of average scaling like 1/N
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
```

```python
def avg(n):
    return sum([ss.norm.rvs() for _ in range(n)])/n
```

```python
n = 10  # number of variates to average
plt.style.use("phys267_jupyter_overhead.mplstyle")
fig, ax = plt.subplots()
ax.set_xlim([-5,5])
ax.set_ylim([0,1.5])
# historgram random variates:
ax.hist([ss.norm.rvs() for i in range(10000)], bins=100, alpha=0.2,
        density=True, label="sampled parent");
# histogram mean of sets of random variates:
ax.hist([avg(n) for i in range(10000)], bins=100, color="red", alpha=0.2,
        density=True, label="sample mean, $N=$"+str(n));
# superimpose theoretical results:
xs = np.linspace(-5, 5, 1000)
ax.plot(xs, ss.norm.pdf(xs), alpha=0.2, color="black", label="expected")
ax.plot(xs, ss.norm.pdf(xs,scale=1.0/np.sqrt(n)), alpha=0.2, color="black")
ax.set_xlabel("$x$")
ax.set_ylabel("$f(x)$")
ax.legend()
```
