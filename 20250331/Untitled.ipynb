```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("highest_fig_2_of_lamb_fine_structure_of_the_hydrogen_shortdoi_cgbz6x.csv")
```

```python
bs = df["b"].to_numpy()
fs = df["f"].to_numpy()
```

```python
def fit(xs, ys):
    n = len(xs)
    assert len(xs) == len(ys)
    t = n * sum(xs * xs) - sum(xs)**2
    m = (n * sum(xs * ys) - sum(xs) * sum(ys)) / t
    b = (sum(xs * xs) * sum(ys) - sum(xs) * sum(xs * ys)) / t
    return m, b
```

```python
m, b = fit(fs, bs)
```

```python
m, b
ffs = np.linspace(9e3,13e3,1000)
print(-b/m)
```

```python
plt.plot(fs, bs, "+")
plt.plot(ffs, m*ffs+b)
ax = plt.gca()
ax.set_xlabel("frequency (MHz)")
ax.set_ylabel("magnetic field (G)")
ax.axhline(0,0,1)
```
