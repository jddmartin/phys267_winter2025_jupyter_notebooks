```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
help(random.random)
```

```python
n = 10**6
variates = [np.sin(2*np.pi*random.random()) for _ in range(n)]
```

```python
plt.style.use("phys267_jupyter_overhead.mplstyle")
fig, ax = plt.subplots()
ax.hist(variates, bins=100, density=True, alpha=0.1, color="green" )
ax.hist(variates, bins=1000, density=True, alpha=0.1, color="red" )
ax.set_ylabel("f")
ax.set_xlabel("sin(x)")
ax.set_ylim([0,2])
ax.set_xlim([-1,1])
#ax.set_xscale("log")  # make x-axis "log"
fig.tight_layout()
fig.savefig("hist_of_sin.pdf")
```

```python
n = 10**5; nu = 50;
variates = [sum([-0.5+random.random() for _ in range(nu)])/nu for _ in range(n)]
```
