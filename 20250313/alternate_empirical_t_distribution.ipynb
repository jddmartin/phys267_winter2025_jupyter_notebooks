```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
def rs(n_samp, n):
    rvs = []
    for _ in range(n_samp):
        samples = ss.norm.rvs(size=n)
        mean_est = sum(samples)/n
        rvs.append(np.sqrt(n)*(mean_est)/(np.sqrt(sum((samples-mean_est)**2)/(n-1))))
    return rvs
```

```python
dof = 2
brvs_more = rs(10000, dof)
```

```python
plt.hist(brvs_more, bins=np.linspace(-10,10,100), alpha=0.3, density=True, label="N=" + str(dof));
xs = np.linspace(-5,5,1000)
#plt.plot(xs, 1.0/np.sqrt(2*np.pi) * np.exp(-0.5 * xs**2))
#plt.plot(xs, ss.t.pdf(xs, df=dof-1),label="-1")
plt.plot(xs, ss.t.pdf(xs, df=dof-1),label="0")
#plt.plot(xs, ss.t.pdf(xs, df=dof+1),label="+1")
plt.legend()
ax = plt.gca()
ax.set_xlim([-10,10])
```
