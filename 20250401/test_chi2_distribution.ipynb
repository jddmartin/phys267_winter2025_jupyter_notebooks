```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def chi2_sv(nu):
    xs = ss.norm.rvs(size=nu)
    return sum(xs * xs)
```

```python
nu = 3
npts = 10000
svs = np.array([chi2_sv(nu) for _ in range(npts)])
mu = np.mean(svs)
var = (1/(npts - 1)) * sum((svs - mu)**2)
print(mu, var)
def chi2_g(chi2):
    return 1/2**(nu/2) * np.sqrt(chi2)**(nu-2) / math.gamma(nu/2) * np.exp(-0.5 * chi2)
xs = np.linspace(0, 3*nu, 10000)
gs = chi2_g(xs)
```

```python
plt.hist(svs,bins=100,density=True);
plt.plot(xs, gs)
```
