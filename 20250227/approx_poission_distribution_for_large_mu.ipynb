```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Poisson distribution:
def f(mu, k):
    return 1.0/math.factorial(k) * mu**k * np.exp(-mu)
```

```python
# Gaussian approximation to Poisson:
def f_approx(mu, k):
    sigma = np.sqrt(mu)
    return 1.0/np.sqrt(2*np.pi)/sigma * np.exp(-0.5*(k-mu)**2/sigma**2)
```

```python
def g(mu):
    ks = list(range(0,100))
    fs = np.array([f(mu,k) for k in ks])
    plt.bar(ks, fs, label="exact",alpha=0.4)
    plt.plot(ks, [f_approx(mu, k) for k in ks], "+", label="approx",color="black")
    ax = plt.gca()
    ax.set_ylabel("f(i)")
    ax.set_xlabel("i")
    ax.set_xlim([-1,10*mu])
#ax.set_yscale("log")
#ax.set_ylim([0.001,1])
#ax.axvline(mu,0,1)
    ax.legend()
```

```python
g(1)
```
