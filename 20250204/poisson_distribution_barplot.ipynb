```python
# Plot Poisson probability distribution using a barplot
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
ks = list(range(100))
mu = 10
fs = [mu**k / math.factorial(k) * np.exp(-mu) for k in ks]
```

```python
plt.bar(ks, fs)
ax = plt.gca()
ax.set_xlim([0,20])
ax.xaxis.get_major_locator().set_params(integer=True) 
ax.minorticks_on()
ax.tick_params(axis='y', which='minor')
ax.set_xlabel("counts, $k$")
ax.set_ylabel("$f(k)$")
```
